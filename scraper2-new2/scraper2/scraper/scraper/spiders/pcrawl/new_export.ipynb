```python
import pandas as pd
import numpy as np
df = pd.read_csv('export3.csv',encoding='utf-8')
df['zipcode'] = df['zipcode'].fillna(0).astype(str)
df['city'] = df['city'].replace(' ', '-',regex=True).str.lower()
df['state'] = df['state'].str.lower()
df
```

```python
df['budget_categories'] = 0
df['low'] = 0
df['high'] = 0
for i in range (0,len(df)):
  if df['budget'][i] == '$399 Or Less':
    df['budget_categories'][i] = 1
    df['low'][i] = 399 * 0.8
    df['high'][i] = 399 * 1.2 
  elif df['budget'][i] == '$400 - $699':
    df['budget_categories'][i] = 2
    df['low'][i] = 400 * 0.8
    df['high'][i] = 699 * 1.2
  elif df['budget'][i] == '$700 - $999':
    df['budget_categories'][i] = 3
    df['low'][i] = 700 * 0.8
    df['high'][i] = 999 * 1.2
  elif df['budget'][i] == '$1000+':
    df['budget_categories'][i] = 4
    df['low'][i] = 1000 
    df['high'][i] = 1000 * 1.2
  else:
    df['budget_categories'][i] = 5
    df['low'][i] = 400
    df['high'][i] = 1000 
df
```
